**Task:** Classify car **brand only** (e.g. BMW, Toyota, Ford)  
**Expected:** ~80-120 classes, ~85-95% accuracy  

This is a **fallback classifier** — when the full make+model classifier

```python
# ============================================================
# CELL 3: Extract VMMRdb Dataset
# ============================================================
# Reuses extracted data if model classifier notebook already ran.

import os
import zipfile
import time

LOCAL_DATA = '/content/data'
RAW_DIR = f'{LOCAL_DATA}/_vmmrdb_raw'
FLAT_DIR = f'{LOCAL_DATA}/vmmrdb'
os.makedirs(LOCAL_DATA, exist_ok=True)


def count_images(root):
    count = 0
    if not os.path.exists(root):
        return 0
    for _, _, files in os.walk(root):
        count += sum(1 for f in files if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.webp')))
    return count


# Check if data already exists (from model classifier notebook)
if os.path.exists(FLAT_DIR) and count_images(FLAT_DIR) > 100:
    print(f'VMMRdb already extracted: {count_images(FLAT_DIR):,} images in {FLAT_DIR}')
elif os.path.exists(RAW_DIR) and count_images(RAW_DIR) > 100:
    print(f'Raw VMMRdb found: {count_images(RAW_DIR):,} images in {RAW_DIR}')
else:
    # Find and extract ZIP
    drive_data = '/content/drive/MyDrive/CoreVision/data'
    DRIVE_ZIP = None
    if os.path.exists(drive_data):
        for f in os.listdir(drive_data):
            if f.lower().endswith('.zip') and 'vmmr' in f.lower():
                DRIVE_ZIP = os.path.join(drive_data, f)
                break
        if DRIVE_ZIP is None:
            for f in os.listdir(drive_data):
                if f.lower().endswith('.zip') and 'car' not in f.lower():
                    DRIVE_ZIP = os.path.join(drive_data, f)
                    break

    if DRIVE_ZIP is None:
        raise FileNotFoundError(f'No VMMRdb ZIP found in {drive_data}/')

    print(f'Extracting {os.path.basename(DRIVE_ZIP)}...')
    t0 = time.time()
    os.makedirs(RAW_DIR, exist_ok=True)
    with zipfile.ZipFile(DRIVE_ZIP, 'r') as zf:
        zf.extractall(RAW_DIR)
    print(f'Extracted in {time.time() - t0:.0f}s')

print(f'VMMRdb images available: {count_images(RAW_DIR if os.path.exists(RAW_DIR) else FLAT_DIR):,}')
print('Cell 3 complete.')
```

```python
# ============================================================
# CELL 4: Brand-Level Dataset Preparation (VMMRdb + Car-1000)
# ============================================================
# VMMRdb dirs: acura_cl_1997, bmw_3_series_2015, etc.
#   Brand = first word (before first underscore that starts a model name)
# Car-1000: uses name_from_old field which has English after ==
#   e.g. '宝马_宝马M5==BMW_M5' -> English brand is 'BMW'
# All models and years under a brand get merged from BOTH datasets.

import os
import re
import json
import random
import shutil
import unicodedata
from collections import defaultdict

RAW_DIR = '/content/data/_vmmrdb_raw'
SPLIT_ROOT = '/content/data/brand_split'
MIN_SAMPLES = 50
MANUAL_BRANDS_DIR = '/content/drive/MyDrive/CoreVision/data/manual_brands'
MANUAL_BRAND_MIN_SAMPLES = 10
MIN_SAMPLES_BY_BRAND = {
    'opel': MANUAL_BRAND_MIN_SAMPLES,
    'seat': MANUAL_BRAND_MIN_SAMPLES,
    'cupra': MANUAL_BRAND_MIN_SAMPLES,
}
VAL_RATIO = 0.2
FORCE_REBUILD = True


def is_image(name):
    return name.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.webp'))


def extract_brand_vmmrdb(dir_name):
    """Extract brand from VMMRdb dir name.
    acura_cl_1997 -> acura
    bmw_3_series_2015 -> bmw
    mercedes-benz_c-class_2014 -> mercedes-benz
    land_rover_range_rover_2010 -> land rover
    """
    # Known multi-word brands (lowercase, underscore-separated)
    MULTI_WORD = {
        'aston_martin', 'alfa_romeo', 'am_general',
        'land_rover', 'range_rover',
        'rolls_royce', 'scion_fr',
    }
    name_lower = dir_name.lower()

    # Check multi-word brands first
    for multi in MULTI_WORD:
        if name_lower.startswith(multi + '_'):
            return multi.replace('_', ' ')

    # Default: first word before underscore
    parts = dir_name.split('_')
    return parts[0].lower()


def extract_brand_car1000(name_from_old):
    """Extract ENGLISH brand from Car-1000 name_from_old field.
    The field format is: 'Chinese_Chinese==EnglishBrand_EnglishModel'
    e.g. '宝马_宝马M5==BMW_M5' -> 'bmw'
         '奔驰_奔驰G级==Mercedes-Benz_Mercedes-Benz G-Class' -> 'mercedes-benz'
         '比亚迪_海豚==BYD_Dolphin' -> 'byd'
         '阿斯顿·马丁_DB12==Aston Martin_DB12' -> 'aston martin'
    """
    # Split on == to get English portion
    if '==' in name_from_old:
        english_part = name_from_old.split('==', 1)[1]  # e.g. 'BMW_M5'
    else:
        english_part = name_from_old

    # Brand is before the first underscore: 'BMW_M5' -> 'BMW'
    brand = english_part.split('_', 1)[0].strip().lower()

    # Handle multi-word brands (brand part before _ may contain spaces)
    MULTI_WORD = {
        'aston martin', 'alfa romeo', 'am general',
        'land rover', 'rolls royce', 'great wall',
        'gac trumpchi', 'saic maxus', 'baic changhe',
        'baic manufacturing', 'baic weiwang',
        'changan kaicheng', 'changan kuayue', 'changan oushang',
        'dongfeng fengshen', 'dongfeng xiaokang',
        'farizon auto', 'jac motors',
        'nezha auto', 'ruilan auto',
        'wuling motors', 'geely galaxy',
        'wei brand',
    }
    for multi in MULTI_WORD:
        if brand == multi or brand.startswith(multi):
            return multi

    return brand


# ---- Normalize brand names across datasets ----
# Map sub-brands and aliases to canonical English brand names
BRAND_ALIASES = {
    'range rover': 'land rover',
    'vw': 'volkswagen',
    'chevy': 'chevrolet',
    'merc': 'mercedes-benz',
    'gac trumpchi': 'gac',
    'saic maxus': 'maxus',
    'baic changhe': 'baic',
    'baic manufacturing': 'baic',
    'baic weiwang': 'baic',
    'changan kaicheng': 'changan',
    'changan kuayue': 'changan',
    'changan oushang': 'changan',
    'dongfeng fengshen': 'dongfeng',
    'dongfeng xiaokang': 'dongfeng',
    'farizon auto': 'farizon',
    'jac motors': 'jac',
    'nezha auto': 'nezha',
    'ruilan auto': 'ruilan',
    'wuling motors': 'wuling',
    'geely galaxy': 'geely',
    'wei brand': 'wey',
}

def normalize_brand(brand):
    brand = brand.strip().lower().replace('-', ' ')
    brand = ''.join(
        c for c in unicodedata.normalize('NFKD', brand)
        if not unicodedata.combining(c)
    )
    brand = ' '.join(brand.split())
    return BRAND_ALIASES.get(brand, brand)


train_dir = f'{SPLIT_ROOT}/train'
val_dir = f'{SPLIT_ROOT}/val'

need_rebuild = FORCE_REBUILD or not os.path.exists(train_dir)

if need_rebuild:
    random.seed(42)
    if os.path.exists(SPLIT_ROOT):
        shutil.rmtree(SPLIT_ROOT)
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(val_dir, exist_ok=True)

    # ==========================================================
    # Part A: Gather images from VMMRdb
    # ==========================================================
    brand_images = defaultdict(list)  # brand -> [(src_path, unique_name)]

    if os.path.exists(RAW_DIR):
        all_dirs = sorted([
            d for d in os.listdir(RAW_DIR)
            if os.path.isdir(os.path.join(RAW_DIR, d))
            and not d.startswith('_')
        ])
        print(f'VMMRdb: Found {len(all_dirs)} raw class directories')

        for dir_name in all_dirs:
            dir_path = os.path.join(RAW_DIR, dir_name)
            brand = normalize_brand(extract_brand_vmmrdb(dir_name))

            for f in os.listdir(dir_path):
                fpath = os.path.join(dir_path, f)
                if is_image(f) and os.path.isfile(fpath):
                    unique = f'vmmr__{dir_name}__{f}'
                    brand_images[brand].append((fpath, unique))

        print(f'VMMRdb: {sum(len(v) for v in brand_images.values()):,} images across {len(brand_images)} brands')
    else:
        print(f'WARNING: VMMRdb not found at {RAW_DIR}')

    # ==========================================================
    # Part B: Gather images from Car-1000
    # ==========================================================
    car1000_count = 0
    car1000_brands_found = set()

    # car1000_images_dir was set in Cell 3.5
    if 'car1000_images_dir' in dir() and car1000_images_dir is not None:
        # Load class_info.json to map category IDs to English brand names
        # name_from_old field has format: 'Chinese==EnglishBrand_Model'
        class_info_path = os.path.join(car1000_images_dir, 'cls_info', 'class_info.json')
        if not os.path.exists(class_info_path):
            # Try alternative path
            for root, dirs, files in os.walk(car1000_images_dir):
                for f in files:
                    if f == 'class_info.json':
                        class_info_path = os.path.join(root, f)
                        break

        if os.path.exists(class_info_path):
            with open(class_info_path, 'r', encoding='utf-8') as f:
                class_info = json.load(f)

            # Build mapping: category_id (str) -> English brand name
            # We use name_from_old which contains English after ==
            # e.g. '宝马_宝马M5==BMW_M5' -> brand='bmw'
            cat_to_brand = {}
            if isinstance(class_info, list):
                for entry in class_info:
                    cat_id = str(entry.get('id', '')).zfill(4)
                    name_old = entry.get('name_from_old', '')
                    if name_old and '==' in name_old:
                        brand = normalize_brand(extract_brand_car1000(name_old))
                        cat_to_brand[cat_id] = brand
            elif isinstance(class_info, dict):
                for cat_id, entry in class_info.items():
                    cat_id_str = str(cat_id).zfill(4)
                    if isinstance(entry, dict):
                        name_old = entry.get('name_from_old', '')
                    else:
                        name_old = str(entry)
                    if name_old and '==' in name_old:
                        brand = normalize_brand(extract_brand_car1000(name_old))
                        cat_to_brand[cat_id_str] = brand

            # Show sample mappings for verification
            print(f'Car-1000: Loaded {len(cat_to_brand)} category-to-brand mappings')
            sample = list(cat_to_brand.items())[:15]
            print(f'  Sample: {sample}')
            unique_brands = sorted(set(cat_to_brand.values()))
            print(f'  Unique brands: {len(unique_brands)}')
            print(f'  Brands: {unique_brands}')

            # Scan all_images/ directories
            all_images_root = os.path.join(car1000_images_dir, 'all_images')
            for cat_dir in sorted(os.listdir(all_images_root)):
                cat_path = os.path.join(all_images_root, cat_dir)
                if not os.path.isdir(cat_path):
                    continue

                cat_id = cat_dir.zfill(4)
                brand = cat_to_brand.get(cat_id)

                if brand is None:
                    # Fallback: try without zero-padding
                    brand = cat_to_brand.get(cat_dir)
                if brand is None:
                    # Skip unknown categories
                    continue

                car1000_brands_found.add(brand)
                for f in os.listdir(cat_path):
                    fpath = os.path.join(cat_path, f)
                    if is_image(f) and os.path.isfile(fpath):
                        unique = f'car1k__{cat_dir}__{f}'
                        brand_images[brand].append((fpath, unique))
                        car1000_count += 1

            print(f'Car-1000: {car1000_count:,} images across {len(car1000_brands_found)} brands')
        else:
            print(f'WARNING: class_info.json not found in Car-1000 data.')
    else:
        print('Car-1000 not available, using VMMRdb only.')

    # ==========================================================
    # Part C: Gather images from Google Drive manual_brands/ (recursive)
    # ==========================================================
    manual_count = 0
    manual_brands_found = set()
    if os.path.exists(MANUAL_BRANDS_DIR):
        for brand_dir in sorted(os.listdir(MANUAL_BRANDS_DIR)):
            brand_path = os.path.join(MANUAL_BRANDS_DIR, brand_dir)
            if not os.path.isdir(brand_path):
                continue

            brand = normalize_brand(brand_dir)
            added_for_brand = 0
            # Supports both flat and nested layouts:
            #   manual_brands/<brand>/*.jpg
            #   manual_brands/<brand>/<model>/*.jpg
            for root, _, files in os.walk(brand_path):
                rel_dir = os.path.relpath(root, brand_path)
                rel_tag = rel_dir.replace(os.sep, '__') if rel_dir != '.' else 'root'

                for f in files:
                    if not is_image(f):
                        continue
                    fpath = os.path.join(root, f)
                    if not os.path.isfile(fpath):
                        continue

                    unique = f'manual__{brand_dir}__{rel_tag}__{f}'
                    brand_images[brand].append((fpath, unique))
                    manual_count += 1
                    added_for_brand += 1

            if added_for_brand > 0:
                manual_brands_found.add(brand)

        print(f'Manual brands: {manual_count:,} images across {len(manual_brands_found)} brands')
    else:
        print(f'Manual brands folder not found (optional): {MANUAL_BRANDS_DIR}')

    # ==========================================================
    # Part D: Merge & Split
    # ==========================================================
    print(f'\nCombined: {len(brand_images)} unique brands, '
          f'{sum(len(v) for v in brand_images.values()):,} total images')

    # Show brand distribution
    top = sorted(brand_images.items(), key=lambda x: -len(x[1]))[:30]
    print(f'\nTop 30 brands:')
    for brand, imgs in top:
        print(f'  {brand}: {len(imgs):,} images')

    # Filter + split
    total_train = 0
    total_val = 0
    valid = 0
    skipped = 0
    skipped_details = []

    for brand in sorted(brand_images.keys()):
        images = brand_images[brand]
        min_required = MIN_SAMPLES_BY_BRAND.get(
            brand,
            MANUAL_BRAND_MIN_SAMPLES if brand in manual_brands_found else MIN_SAMPLES
        )
        if len(images) < min_required:
            skipped += 1
            skipped_details.append((brand, len(images), min_required))
            continue

        random.shuffle(images)
        split_idx = max(1, min(int(len(images) * (1 - VAL_RATIO)), len(images) - 1))
        train_imgs = images[:split_idx]
        val_imgs = images[split_idx:]

        os.makedirs(os.path.join(train_dir, brand), exist_ok=True)
        os.makedirs(os.path.join(val_dir, brand), exist_ok=True)

        for src, fname in train_imgs:
            shutil.copy2(src, os.path.join(train_dir, brand, fname))
        for src, fname in val_imgs:
            shutil.copy2(src, os.path.join(val_dir, brand, fname))

        total_train += len(train_imgs)
        total_val += len(val_imgs)
        valid += 1

    print(f'\nDataset results:')
    print(f'  Brands found:     {len(brand_images)}')
    print(f'  Skipped (< brand threshold):  {skipped}')
    if skipped_details:
        print('  Sample skipped brands:')
        for brand, n_img, min_req in skipped_details[:20]:
            print(f'    - {brand}: {n_img} (needs >= {min_req})')
    print(f'  Valid brands:     {valid}')
    print(f'  Train images:     {total_train:,}')
    print(f'  Val images:       {total_val:,}')
    print(f'  Avg imgs/brand:   {(total_train + total_val) // max(valid, 1)}')
else:
    print('Existing brand split found.')

classes = sorted([c for c in os.listdir(train_dir)
                  if os.path.isdir(os.path.join(train_dir, c))])
n_classes = len(classes)
print(f'\nFinal: {n_classes} brand classes')
print(f'Brands: {classes}')

class_names_path = '/content/drive/MyDrive/CoreVision/data/vmmrdb_brand_classes.txt'
with open(class_names_path, 'w') as f:
    f.write('\n'.join(classes))
print(f'Saved to Drive.')
```

```python
# ============================================================
# CELL 5: Hyperparameters
# ============================================================
import os

SPLIT_ROOT = '/content/data/brand_split'

CONFIG = {
    'train_dir': f'{SPLIT_ROOT}/train',
    'val_dir': f'{SPLIT_ROOT}/val',
    'num_classes': n_classes,
    'img_size': 224,
    'batch_size': 256,
    'num_epochs': 15,
    'freeze_epochs': 3,
    'num_workers': 4,
    'lr_backbone': 2e-4,
    'lr_head': 1e-3,
    'weight_decay': 1e-4,
    'save_path': '/content/drive/MyDrive/CoreVision/weights/brand_classifier.pth',
    'resume_from': None,
}

print('Configuration:')
for k, v in CONFIG.items():
    print(f'  {k}: {v}')
print(f'\nPhase 1: Epochs 1-{CONFIG["freeze_epochs"]} (frozen)')
print(f'Phase 2: Epochs {CONFIG["freeze_epochs"]+1}-{CONFIG["num_epochs"]} (fine-tune)')
```

```python
# ============================================================
# CELL 6: Data Loaders
# ============================================================
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

IMG_SIZE = CONFIG['img_size']

train_transforms = transforms.Compose([
    transforms.RandomResizedCrop(IMG_SIZE, scale=(0.7, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

val_transforms = transforms.Compose([
    transforms.Resize(int(IMG_SIZE * 1.1)),
    transforms.CenterCrop(IMG_SIZE),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

train_dataset = datasets.ImageFolder(CONFIG['train_dir'], transform=train_transforms)
val_dataset = datasets.ImageFolder(CONFIG['val_dir'], transform=val_transforms)

if len(train_dataset) == 0 or len(val_dataset) == 0:
    raise RuntimeError(
        f'Empty dataset split. train={len(train_dataset)}, val={len(val_dataset)}. '
        f'Check dataset extraction and MIN_SAMPLES settings.'
    )

if len(train_dataset.classes) < 2:
    raise RuntimeError(
        f'Need at least 2 classes for classification, found {len(train_dataset.classes)}.'
    )

effective_train_bs = min(CONFIG['batch_size'], max(1, len(train_dataset)))
effective_val_bs = min(CONFIG['batch_size'], max(1, len(val_dataset)))
if effective_train_bs != CONFIG['batch_size']:
    print(f'Adjusted train batch size: {CONFIG["batch_size"]} -> {effective_train_bs}')
if effective_val_bs != CONFIG['batch_size']:
    print(f'Adjusted val batch size: {CONFIG["batch_size"]} -> {effective_val_bs}')

persistent = CONFIG['num_workers'] > 0
pin = torch.cuda.is_available()
train_loader = DataLoader(
    train_dataset, batch_size=effective_train_bs,
    shuffle=True, num_workers=CONFIG['num_workers'],
    pin_memory=pin, drop_last=True, persistent_workers=persistent
)
val_loader = DataLoader(
    val_dataset, batch_size=effective_val_bs,
    shuffle=False, num_workers=CONFIG['num_workers'],
    pin_memory=pin, persistent_workers=persistent
)

if len(train_loader) == 0:
    raise RuntimeError(
        f'Train loader has 0 batches. train={len(train_dataset)}, bs={effective_train_bs}. '
        f'Reduce batch_size or check data split.'
    )

print(f'Train: {len(train_dataset)} images, {len(train_loader)} batches')
print(f'Val:   {len(val_dataset)} images, {len(val_loader)} batches')
print(f'Brands: {len(train_dataset.classes)}')
```

```python
# ============================================================
# CELL 7: Build Model
# ============================================================
import torch
import torch.nn as nn
import timm

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')


def build_model(num_classes, pretrained=True):
    backbone = timm.create_model('tf_efficientnetv2_s', pretrained=pretrained, num_classes=0)
    feature_dim = backbone.num_features
    model = nn.Sequential(
        backbone,
        nn.Dropout(0.3),
        nn.Linear(feature_dim, 256),
        nn.BatchNorm1d(256),
        nn.ReLU(inplace=True),
        nn.Dropout(0.2),
        nn.Linear(256, num_classes)
    )
    return model, feature_dim


def _get_raw(model):
    if hasattr(model, '_orig_mod'):
        return model._orig_mod
    return model


def freeze_backbone(model):
    for param in _get_raw(model)[0].parameters():
        param.requires_grad = False
    t = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f'Backbone FROZEN - trainable: {t:,}')


def unfreeze_backbone(model):
    for param in _get_raw(model)[0].parameters():
        param.requires_grad = True
    t = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f'Backbone UNFROZEN - trainable: {t:,}')


def get_backbone_params(model):
    return list(_get_raw(model)[0].parameters())


def get_head_params(model):
    return list(_get_raw(model)[1:].parameters())


model, feature_dim = build_model(CONFIG['num_classes'])
model = model.to(device)
freeze_backbone(model)

total = sum(p.numel() for p in model.parameters())
print(f'Total params: {total:,}')
```

```python
# ============================================================
# CELL 8: Optimizer & Loss
# ============================================================
import torch.optim as optim

optimizer = optim.AdamW(
    [p for p in get_head_params(model) if p.requires_grad],
    lr=CONFIG['lr_head'],
    weight_decay=CONFIG['weight_decay']
)

scheduler = optim.lr_scheduler.CosineAnnealingLR(
    optimizer, T_max=CONFIG['freeze_epochs'], eta_min=1e-5
)

criterion = nn.CrossEntropyLoss(label_smoothing=0.1)

print('Optimizer ready (head-only for Phase 1)')
```

```python
# ============================================================
# CELL 9: Training Loop
# ============================================================
import time

start_epoch = 0
best_val_acc = 0.0
history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}
amp_enabled = device.type == 'cuda'
scaler = torch.amp.GradScaler('cuda', enabled=amp_enabled)

SAVE_DIR = os.path.dirname(CONFIG['save_path'])
LATEST_PATH = os.path.join(SAVE_DIR, 'brand_classifier_latest.pth')


def save_checkpoint(path, epoch, val_acc, val_top5=None, is_best=False):
    ckpt = {
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'accuracy': val_acc,
        'top5_accuracy': val_top5,
        'num_classes': CONFIG['num_classes'],
        'class_names': train_dataset.classes,
        'history': history,
        'best_val_acc': best_val_acc,
        'classifier_type': 'brand',
    }
    torch.save(ckpt, path)
    tag = 'BEST' if is_best else 'latest'
    print(f'  [{tag}] {os.path.basename(path)} (val_acc={val_acc:.2%})')


# Resume
resume_path = CONFIG['resume_from']
if not resume_path or not os.path.exists(str(resume_path or '')):
    if os.path.exists(LATEST_PATH):
        resume_path = LATEST_PATH
    elif os.path.exists(CONFIG['save_path']):
        resume_path = CONFIG['save_path']

if resume_path and os.path.exists(str(resume_path)):
    _ckpt = torch.load(resume_path, map_location='cpu')
    ckpt_classes = _ckpt.get('num_classes') if isinstance(_ckpt, dict) else None
    ckpt_class_names = _ckpt.get('class_names') if isinstance(_ckpt, dict) else None
    old_state = _ckpt.get('model_state_dict') if isinstance(_ckpt, dict) else None
    if old_state is None and isinstance(_ckpt, dict):
        old_state = {k: v for k, v in _ckpt.items() if hasattr(v, 'shape')}

    same_class_count = ckpt_classes == CONFIG['num_classes']
    same_class_order = list(ckpt_class_names or []) == list(train_dataset.classes)

    if same_class_count and same_class_order and old_state is not None:
        model.load_state_dict(old_state)
        start_epoch = _ckpt.get('epoch', -1) + 1
        best_val_acc = _ckpt.get('best_val_acc', 0.0)
        history = _ckpt.get('history', history)
        if start_epoch >= CONFIG['freeze_epochs']:
            unfreeze_backbone(model)
            optimizer = optim.AdamW([
                {'params': get_backbone_params(model), 'lr': CONFIG['lr_backbone']},
                {'params': get_head_params(model), 'lr': CONFIG['lr_head'] * 0.3},
            ], weight_decay=CONFIG['weight_decay'])
            ft = CONFIG['num_epochs'] - CONFIG['freeze_epochs']
            scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=ft, eta_min=1e-6)
        print(f'Resumed from epoch {start_epoch}, best: {best_val_acc:.2%}')
    elif old_state is not None:
        current_state = model.state_dict()
        transferable = {}
        for k, v in old_state.items():
            if k not in current_state:
                continue
            if tuple(v.shape) != tuple(current_state[k].shape):
                continue
            # Keep classifier layer randomly initialized when class mapping changed.
            if k.endswith('6.weight') or k.endswith('6.bias'):
                continue
            transferable[k] = v
        model.load_state_dict(transferable, strict=False)
        print(f'Checkpoint not directly resumable (classes/order changed or mismatch).')
        print(f'Warm-started {len(transferable)} tensors; classifier head is reinitialized.')
    else:
        print('Checkpoint format not recognized. Starting fresh.')
    del _ckpt


def train_epoch(model, loader, optimizer, criterion, scaler, device, ep, total):
    model.train()
    loss_sum, correct, count = 0.0, 0, 0
    for i, (imgs, labels) in enumerate(loader):
        imgs = imgs.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)
        optimizer.zero_grad(set_to_none=True)
        with torch.amp.autocast(device_type=device.type, enabled=amp_enabled):
            out = model(imgs)
            loss = criterion(out, labels)
        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        scaler.step(optimizer)
        scaler.update()
        correct += (out.argmax(1) == labels).sum().item()
        loss_sum += loss.item() * imgs.size(0)
        count += imgs.size(0)
        if i % 50 == 0:
            print(f'  Batch {i+1}/{len(loader)} | Loss: {loss.item():.4f}', end='\r')
    return loss_sum / count, correct / count


@torch.no_grad()
def val_epoch(model, loader, criterion, device):
    model.eval()
    loss_sum, correct, top5_correct, count = 0.0, 0, 0, 0
    for imgs, labels in loader:
        imgs = imgs.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)
        with torch.amp.autocast(device_type=device.type, enabled=amp_enabled):
            out = model(imgs)
            loss = criterion(out, labels)
        correct += (out.argmax(1) == labels).sum().item()
        _, t5 = out.topk(min(5, out.size(1)), dim=1)
        top5_correct += t5.eq(labels.unsqueeze(1)).any(1).sum().item()
        loss_sum += loss.item() * imgs.size(0)
        count += imgs.size(0)
    return loss_sum / count, correct / count, top5_correct / count


FREEZE = CONFIG['freeze_epochs']
TOTAL = CONFIG['num_epochs']
phase2 = start_epoch >= FREEZE

print(f'Training {TOTAL} epochs ({FREEZE} frozen + {TOTAL-FREEZE} fine-tune)')
print('=' * 60)

for epoch in range(start_epoch, TOTAL):
    t0 = time.time()

    if epoch == FREEZE and not phase2:
        phase2 = True
        print('\n' + '=' * 60)
        print('PHASE 2: Unfreezing backbone')
        print('=' * 60)
        unfreeze_backbone(model)
        optimizer = optim.AdamW([
            {'params': get_backbone_params(model), 'lr': CONFIG['lr_backbone']},
            {'params': get_head_params(model), 'lr': CONFIG['lr_head'] * 0.3},
        ], weight_decay=CONFIG['weight_decay'])
        scheduler = optim.lr_scheduler.CosineAnnealingLR(
            optimizer, T_max=TOTAL - FREEZE, eta_min=1e-6)
        scaler = torch.amp.GradScaler('cuda', enabled=amp_enabled)

    tag = 'frozen' if epoch < FREEZE else 'fine-tune'
    print(f'\nEpoch {epoch+1}/{TOTAL} [{tag}]')

    tl, ta = train_epoch(model, train_loader, optimizer, criterion, scaler, device, epoch, TOTAL)
    vl, va, v5 = val_epoch(model, val_loader, criterion, device)
    scheduler.step()

    history['train_loss'].append(tl)
    history['train_acc'].append(ta)
    history['val_loss'].append(vl)
    history['val_acc'].append(va)

    elapsed = time.time() - t0
    print(f'Epoch {epoch+1:02d}/{TOTAL} [{tag}] '
          f'| Train: {tl:.4f} {ta:.2%} '
          f'| Val: {vl:.4f} {va:.2%} Top5: {v5:.2%} '
          f'| {elapsed:.0f}s')

    if va > best_val_acc:
        best_val_acc = va
        save_checkpoint(CONFIG['save_path'], epoch, va, v5, is_best=True)
    save_checkpoint(LATEST_PATH, epoch, va, v5, is_best=False)

print(f'\nDone! Best val accuracy: {best_val_acc:.2%}')
```

```python
# ============================================================
# CELL 10: Plot Training History
# ============================================================
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
epochs = range(1, len(history['train_loss']) + 1)
freeze_ep = CONFIG.get('freeze_epochs', 3)

axes[0].plot(epochs, history['train_loss'], 'b-o', label='Train', ms=3)
axes[0].plot(epochs, history['val_loss'], 'r-o', label='Val', ms=3)
axes[0].axvline(x=freeze_ep + 0.5, color='green', ls='--', alpha=0.7, label='Unfreeze')
axes[0].set_title('Loss'); axes[0].set_xlabel('Epoch'); axes[0].legend(); axes[0].grid(True)

axes[1].plot(epochs, [a*100 for a in history['train_acc']], 'b-o', label='Train', ms=3)
axes[1].plot(epochs, [a*100 for a in history['val_acc']], 'r-o', label='Val', ms=3)
axes[1].axvline(x=freeze_ep + 0.5, color='green', ls='--', alpha=0.7, label='Unfreeze')
axes[1].set_title('Accuracy (%)'); axes[1].set_xlabel('Epoch'); axes[1].legend(); axes[1].grid(True)

plt.suptitle(f'Brand Classifier (VMMRdb + Car-1000) | Best: {best_val_acc:.2%}', fontsize=13)
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/CoreVision/brand_training_history.png', dpi=150)
plt.show()
```
